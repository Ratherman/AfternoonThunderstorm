# 看看我們的 Data

```python
import pandas as pd
df = pd.read_csv("./data/all_data_with_label_v2.csv", index_col=0)
df.head()
```

# 來用 scikit-learn 看看

```python
import sklearn
from sklearn import svm, preprocessing

df = sklearn.utils.shuffle(df)
df=df.fillna(df.mean())
X = df.drop({"Label118","Name117"}, axis=1).values
#X = preprocessing.scale(X)

df["Label118"] = df["Label118"].map({True:1, False:0})
y = df["Label118"].values

test_size = 100

X_train = X[:-test_size]
y_train = y[:-test_size]

X_test = X[-test_size:]
y_test = y[-test_size:]

clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
